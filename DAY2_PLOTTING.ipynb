```python
'''there ATR 5 subject psychology,nanotechnology,medicine,oncology,and antrapology.Two students have applied to do masters in any of tgse five fields with the line plot .show in which field is excel.the cut off for each subject is 70.
Std1=[67,45,73,71,57]
Std2=[43,67,68,53,85]'''
sub=["psy","nano","med","onc","anthro"]
std1=[67,45,73,71,57]
std2=[43,67,68,53,85]
cut=[70,70,70,70,70]
#plt.plot(x,y,label,color,marker,linewidth,linestyle)
plt.scatter(sub,std1,label="student1",color="k",marker="P")
plt.scatter(sub,std2,label="student2",color="b",marker="H")
plt.plot(sub,cut,label="cutoff")
plt.legend() #indicates what line in the graph
plt.title("student eligibility graph")
plt.xlabel("subjects")
plt.ylabel("marks")
```

```python
'''An India vs Australia match online .India has given 298 as target.the runs for every 5 overs as given as follows
India for every 5 overs:-25,51,84,131,160,189,220,250,267,297
Australia score has completed 25 overs:- 15,41,94,110,151,
plot overs vs runs for both the countries with india has blue line and Australia as yellow line with required label with ledgends
The Market india is flag, Australia is plus
Take overs from arange function create runs as arrays'''

from matplotlib import pyplot as plt
import numpy as np
Ind_Overs=np.arange(5,51,5)
Aus_Overs=np.arange(5,26,5)
ind=[25,51,84,131,160,189,220,250,267,297]
aus=[15,41,94,110,151]
plt.plot(Ind_Overs,ind,label="india",color="b",marker=">")
plt.plot(Aus_Overs,aus,label="Australia",color="y",marker="+")
plt.legend()
plt.title("IND VS AUS")
plt.xlabel("overs")
plt.ylabel("score")
plt.grid()
```
